```python
import pandas as pd
```

```python
df = pd.read_excel('data/ILGA_State_Sponsored_Homophobia_2019_dataset.xlsx', sheet_name='All categories')
```

```python
df
```

```python
df.tail(15)
```

```python
for colname in df.columns:
    df.rename(columns={colname : df[colname][1]}, inplace=True)
```

```python
df = df.iloc[2:201].reset_index(drop=True)
```

```python
df.drop(columns={'N', 'CN', 'CONTINENT'}, inplace=True)
```

```python
df
```

```python
df.rename(columns={'COUNTRY':'country', 'Legal?':'legal_male', 'Gender/s':'legal_female', 
                   'Years in Prison / Other':'prison_death', 'Constitutional':'cons_protect', 
                   'Broad Protections':'broad_protect', 'Employment':'employ_protect',
                   'Hate Crime / Aggr. Circs.':'crim_hate', 'Incitement to Hatred / Violence':'crim_incite',
                   'Ban on CTs':'ban_ct', 'Marriage':'marriage', 'Civil Unions':'unions', 'Joint Adoption':'j_adopt', 'Second Parent Adoption':'sec_adopt'},
                    inplace=True)
```

```python
df = df[['country', 'legal_male', 'legal_female', 'prison_death', 'cons_protect',
              'broad_protect', 'employ_protect', 'crim_hate', 'crim_incite', 'ban_ct',
              'marriage', 'unions', 'j_adopt', 'sec_adopt']]
```

```python
for colname in df.columns:
    print(colname)
    print(df[colname].unique())
```

```python
df['legal_female'].replace(to_replace='ANY GENDER', value=0, inplace=True)
df['legal_female'].replace(to_replace='MALE ONLY', value=1, inplace=True)
df['legal_female'].replace(to_replace='M ONLY', value=1, inplace=True)
df['legal_female'].replace(to_replace='DOES NOT APPLY', value=1, inplace=True)
```

```python
df.replace(to_replace={2, 7, 5, 'UNDETERMINED', 3, 14, 'DEATH (NOT)', 'DEATH (REG)', 10, 'DEATH', 'FOR LIFE', 15, 20, 8}, value=1, inplace=True)
df.replace(to_replace='DOES NOT APPLY', value=0, inplace=True)
```

```python
df.replace(to_replace='NO', value=0, inplace=True)
df.replace(to_replace='YES', value=1, inplace=True)
df.replace(to_replace='DE FACTO', value=0, inplace=True)
df.replace(to_replace='N', value=0, inplace=True)
df.replace(to_replace='Y', value=1, inplace=True)
df.replace(to_replace='N ', value=0, inplace=True)
df.replace(to_replace='Y ', value=1, inplace=True)
```

```python
for colname in df.columns:
    df['country'] = df['country'].str.strip()
```
