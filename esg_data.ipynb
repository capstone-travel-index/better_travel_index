```python
col_to_keep = ['iso3', 'ind', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
```

```python
df_esg = souv_esg.loc[:,col_to_keep]
```

```python
col_num = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']

for i in col_num:
    df_esg[i] = pd.to_numeric(df_esg[i], downcast='float', errors='coerce')
```

```python
def print_count(df, country, indic):
    df[df['iso3'] == country & df['ind'] == indic].plot()
```

```python
df_num = df_esg.drop(['iso3', 'ind'], axis=1)
df_norm = (df_num - df_num.min()) / (df_num.max() - df_num.min())
df_norm[['iso3', 'ind']] = df_esg[['iso3', 'ind']]  # Assign the 'iso3' and 'ind' columns from df_esg to df_norm

df_norm.head(10)
```
